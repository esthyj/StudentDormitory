```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('seoul - SeoulRealEstate.csv', engine='python')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="darkgrid")
corr = df.corr()
f, ax = plt.subplots(figsize=(11, 9))
X = sns.heatmap(corr,  cmap='Blues', vmax=1, vmin=-1, center=0,annot=True, fmt='.2f',
                square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
