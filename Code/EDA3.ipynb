```python
import pandas as pd
import numpy as np
```

```python
df0 = pd.read_csv('서울시 대학 및 전문대학 DB 정보 (한국어).csv',engine='python')
df0.head()
```

```python
df1=df0[['학교명','주소','행정구','행정동']]
df1.head()
```

```python
x1 = pd.ExcelFile('2021 고등교육기관 외국인 유학생 현황.xlsx')
df2 = x1.parse('학교별')
df2.head()
```

```python
df3=df2[df2['Unnamed: 5']=='서울']
df3.head()
```

```python
df4=df3[['Unnamed: 2','Unnamed: 8']]
df4.head()
#Unnamed 8 is the number of international students
```

```python
df5= pd.merge(left = df1, right = df4,how= 'inner',left_on='학교명', right_on='Unnamed: 2')
df5.head()
df5.dtypes
```

```python
df5['Unnamed: 8']=pd.to_numeric(df5['Unnamed: 8'])
df5.dtypes
```

```python
# get sum of the international students grouped by 행정동
df5['sum']=df5.groupby(['행정동'])['Unnamed: 8'].transform('sum')
df5.head()
```

```python
df6=df5.sort_values(by='sum',ascending=False)
df6
```

```python
import folium
import json
```

```python
geo_data='seoul_submunicipalities_geo_simple.json'
geo_str=json.load(open(geo_data,encoding='utf-8'))
```

```python
# Seoul center location
center = [37.541, 126.986]
m = folium.Map(location=center, zoom_start=11.2)

folium.Choropleth(
    geo_data=geo_str,
    data=df5,
    columns=('행정동', 'sum'),
    key_on='feature.properties.name',
    fill_color='OrRd',
    legend_name='number of international students',
    tiles='cartodbpositron'
).add_to(m)

m
```

```python
m.save('seoul_map1.html')
m
```
