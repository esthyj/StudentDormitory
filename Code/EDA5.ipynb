```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',None)
```

```python
x1 = pd.ExcelFile('2021 고등교육기관 외국인 유학생 현황.xlsx')
df1 = x1.parse('학교별X국가별')
df1.head()
```

```python
df2=df1[df1['Unnamed: 5']=='서울']
df2.head()
```

```python
df2=df2.astype({'Unnamed: 11':'int'})
```

```python
df2['sum']=df2.groupby(['Unnamed: 8'])['Unnamed: 11'].transform('sum')
```

```python
df3=df2[['Unnamed: 8','sum']]
```

```python
df4=df3.drop_duplicates(['Unnamed: 8'])
```
